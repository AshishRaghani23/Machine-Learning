```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv('Automobile+%281%29.csv')
```

```python
auto.head()
```

```python
sns.displot(auto['normalized_losses'], rug=True)
plt.show()
```

```python
sns.displot(auto['city_mpg'], kde=True, rug=True)
plt.show()
```

```python
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'])
plt.show()
```

```python
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'], kind='hex')
plt.show()
```

```python
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']])
```

```python
sns.stripplot(x=auto['fuel_type'], y=auto['horsepower'])
```

```python
sns.stripplot(x=auto['fuel_type'], y=auto['horsepower'])
```

```python
sns.boxplot(x=auto['number_of_doors'], y=auto['horsepower'], hue=auto['fuel_type'])
```

```python
sns.barplot(x=auto['body_style'], y=auto['horsepower'], hue=auto['engine_location'])
```

```python
sns.pointplot(x=auto['fuel_system'], y=auto['horsepower'], hue=auto['number_of_doors'])
```

```python
sns.catplot(x="fuel_type",
               y="horsepower",
               hue="number_of_doors",
               col="engine_location",
               data=auto,
               kind="strip")
```
